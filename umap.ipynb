## Digits dataset embedding

```python
import numpy as np
import umap
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

digits = load_digits()
X = digits.data
X = StandardScaler().fit_transform(X)
embedding = umap.UMAP(random_state=42).fit_transform(X)

plt.figure(figsize=(6, 5))
plt.scatter(embedding[:, 0], embedding[:, 1], c=digits.target, cmap='Spectral', s=6)
plt.colorbar()
plt.tight_layout()
plt.show()
```

## Random 4D to 2D visualization

```python
import numpy as np
import umap
import matplotlib.pyplot as plt

np.random.seed(44)
data = np.random.rand(700, 4)
embedding = umap.UMAP(random_state=42).fit_transform(data)
colors = data[:, :3]

plt.figure(figsize=(6, 5))
plt.scatter(embedding[:, 0], embedding[:, 1], c=colors, s=6)
plt.tight_layout()
plt.show()
```

## Parameter tuning

```python
import numpy as np
import umap
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

digits = load_digits()
X = StandardScaler().fit_transform(digits.data)
params = [(5, 0.5), (15, 0.1), (50, 0.3)]
embeddings = [umap.UMAP(n_neighbors=n, min_dist=d, random_state=42).fit_transform(X) for n, d in params]

plt.figure(figsize=(12, 4))
for i, (emb, (n, d)) in enumerate(zip(embeddings, params)):
    ax = plt.subplot(1, 3, i + 1)
    ax.scatter(emb[:, 0], emb[:, 1], c=digits.target, cmap='Spectral', s=5)
    ax.set_title(f"n_neighbors={n}, min_dist={d}")
plt.tight_layout()
plt.show()
```

## Supervised UMAP

```python
import numpy as np
import umap
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

digits = load_digits()
X = StandardScaler().fit_transform(digits.data)
y = digits.target
embedding = umap.UMAP(n_neighbors=30, min_dist=0.1, random_state=42).fit_transform(X, y)

plt.figure(figsize=(6, 5))
plt.scatter(embedding[:, 0], embedding[:, 1], c=y, cmap='Spectral', s=6)
plt.colorbar()
plt.tight_layout()
plt.show()
```

## Key parameters
- n_neighbors controls local vs global structure
- min_dist controls cluster compactness
- n_components sets embedding dimensionality
- metric chooses distance function
- random_state for reproducibility
